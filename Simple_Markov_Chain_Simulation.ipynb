```python
# Generating Adjacent Matrix
A  = [[0.2, 0.5, 0.3],
      [0.4, 0.2, 0.4],
      [0.25, 0.45, 0.3]]
```

```python
# Specifying the states
states = ["Home", "Gym", "Cafe"]

# Specifying the the starting state and number of iteration
current_state = "Home"
num_days = 10

# A Dictionary to count number of times a certain action is done
times = dict.fromkeys(states, 0)

for i in range(num_days):
    print(f"Day {i+1}: {current_state}")
    times[current_state]=times.get(current_state)+1
    current_state = np.random.choice(states, p=A[states.index(current_state)])

#Calculating the Stationary Transition 
stat_trans={k:v/num_days for k,v in times.items()}
print(times)
print(stat_trans)
```
